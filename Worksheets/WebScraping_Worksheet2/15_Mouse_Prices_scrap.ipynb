```python
import shutil
import os
import requests
import pandas as pd
import csv


def make_directory(dirname):
    current_path = os.getcwd()
    path = os.path.join(current_path,dirname)
    if not os.path.exists(path):
        os.makedirs(path)

def save_data_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, mode="a", encoding="utf-8-sig")
```

```python
#scrapping images urls using selenium

def scrap_image_url(driver):
    model_names=driver.find_elements_by_xpath("//h2/a[@class='a-link-normal a-text-normal']")
    prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")

    print(len(model_names),len(prices))
    
    product_data={}
    product_data['model_names']=[]
    product_data['prices']=[]
    counter = 0

    for model_name in model_names:
        product_data['model_names'].append(model_name.text)

    for price in prices:
        product_data['prices'].append(price.text)


    print("Returning scrapped data")

    return product_data
        
```

```python

from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

DRIVER_PATH = 'E:\\Softwares\\chromedriver.exe'

driver = webdriver.Chrome(executable_path = DRIVER_PATH)
current_page_url = driver.get('https://www.amazon.in/s?i=computers&bbn=1375420031&rh=n%3A976392031%2Cn%3A976393031%2Cn%3A1375248031%2Cn%3A1375412031%2Cn%3A1375420031%2Cp_36%3A10000-50000%2Cp_85%3A10440599031&dc&qid=1601108830&rnid=10440598031&ref=sr_nr_p_85_1')


start_page =1
total_pages = 1

for page in range(start_page,total_pages+1):
    try:
        product_details = scrap_image_url(driver=driver)
        print("Scrapping Page {0} of {1} pages".format(page,total_pages))

        page_value = driver.find_element_by_xpath("//li[@class='a-selected']").text
        print("The current page scrapped is {}".format(page_value))

        save_data_to_csv(data=product_details, filename='mouse_prices.csv')

    except StaleElementReferenceException as Exception:

        print("We are facing an exception")

```
